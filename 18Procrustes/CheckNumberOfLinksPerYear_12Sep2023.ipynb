```python
import pandas as pd
import numpy as np
from scipy.sparse import csr_matrix, hstack
from scipy.io import mmread, mmwrite
import matplotlib.pyplot as plt
```

```python
A_list = []
for t in range(2005, 2011):
    filename = "A_LAD_" + str(t) + "_GB.mtx"
    A = mmread(filename)
    A_list.append(A)    
```

```python
Asums = []
for t in range(len(A_list)):
    A = A_list[t]
    A = A.todense()
    Asum = np.sum(A)
    Asums.append(Asum)
```

```python
plt.plot(Asums, color="purple", linewidth=2.5)
plt.title("Number of links per year", fontsize=24)
plt.xlabel("Year", fontsize=15)
plt.ylabel("Number of links", fontsize=15)
plt.xticks(range(len(Asums)), range(2005, 2011), fontsize=14)
plt.yticks(fontsize=14)

# save the plot as a pdf file
plt.savefig("NumberOfLinksPerYear.pdf", bbox_inches='tight')
```
